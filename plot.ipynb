```python
import matplotlib.pyplot as plt
import pandas as pd

# 数据构建
data = {
    "Model": [
        "GPT-4o-mini", "GPT-4o", "LLaVa-NeXT-7b", "Qwen2-VL-7b", "LLaVA-OneVision-7b",
        "Intern-VL2-8b", "LLaMa-3.2-11b", "LLaVa-NeXT-13b", "Intern-VL2-40b", "Random Baseline"
    ],
    "2D_Patheval": [0.520, 0.665, 0.501, 0.502, 0.505, 0.489, 0.480, 0.509, 0.506, 0.500],
    "2D_Patheval_w_Desc": [0.750, 0.860, 0.524, 0.731, 0.718, 0.654, 0.695, 0.620, 0.688, 0.500],
    "3D_Patheval": [0.508, 0.624, 0.499, 0.511, 0.509, 0.505, 0.460, 0.494, 0.496, 0.500],
    "3D_Patheval_w_Desc": [0.745, 0.895, 0.517, 0.742, 0.739, 0.691, 0.680, 0.601, 0.717, 0.500],
    "Desc_Only": [0.680, 0.894, 0.514, 0.737, 0.721, 0.648, 0.686, 0.630, 0.679, 0.500]
}

df = pd.DataFrame(data)

# 设置绘图参数
fig, ax = plt.subplots(figsize=(14, 6))
bar_width = 0.15
x = range(len(df))

# 绘制每个系列
for i, col in enumerate(df.columns[1:]):
    ax.bar(
        [pos + i * bar_width for pos in x],
        df[col],
        width=bar_width,
        label=col.replace("_", " ")
    )

# 设置刻度和标签
ax.set_xticks([pos + 2 * bar_width for pos in x])
ax.set_xticklabels(df["Model"], rotation=45, ha='right')
ax.set_ylim(0.45, 1.0)
ax.set_ylabel("Accuracy")
ax.set_title("Model Performance on PaTHEval (2D, 3D, Desc Only)")
ax.legend()

plt.tight_layout()
plt.show()
```
